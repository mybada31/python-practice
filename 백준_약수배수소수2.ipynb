```python
#4
import math
N =int(input())
loc = []
distance = []
for i in range(N):
    a = int(input())
    loc.append(a)
loc.sort()
for i in range(1, N):
    distance.append(abs(loc[i]-loc[i-1]))
count = 0
r = distance[0]
for i in range(len(distance)):
    r = math.gcd(r, distance[i-1])
loc_set = set(loc)
for i in range(loc[0], loc[-1]+1, r):
    if i not in loc_set:
        count+=1
print(count)
```

```python
#7

def is_prime(n):
    if n==1:
        return False
    if n==2:
        return True
    if n%2==0:
        return False
    for i in range(3, int(n**(1/2)+1), 2):
        if n%i==0:
            return False
    return True
def btw_prime(n):
    count = 0
    for i in range(n+1, 2*n+1):
        if is_prime(i):
            count+=1
    return count

while True:
    a = int(input())
    if a==0:
        break
    print(btw_prime(a))
    
    
    
```

```python
#8
T = int(input())

def isprime(n):
    for i in range(2, n):
        if n%i==0:
            return False
    return True
    
def goldbach(n):
    prime = []
    count = 0
    for i in range(2,n):
        if isprime(i)==True:
            prime.append(i)
    length = len(prime)
    for i in range(length-1):
        for j in range(i, length):
            if prime[i]+prime[j]==n:
                count+=1
    return count

for i in range(T):
    a = int(input())
    print(goldbach(a))
    
```
